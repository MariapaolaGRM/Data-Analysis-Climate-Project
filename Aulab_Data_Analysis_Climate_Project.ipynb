```python
# Caricamento dei dataset gas
datasets = {
    'CH4' : r'C:\Users\Mariapaola\Documents\CorsiAulab\Progetto finale\Dataset\emission_dataset(1970-2023)\CH4_Emissions.csv',
    'CO2' : r'C:\Users\Mariapaola\Documents\CorsiAulab\Progetto finale\Dataset\emission_dataset(1970-2023)\CO2_Emissions.csv',
    'GHG' : r'C:\Users\Mariapaola\Documents\CorsiAulab\Progetto finale\Dataset\emission_dataset(1970-2023)\GHG_Emissions.csv',
    'N2O' : r'C:\Users\Mariapaola\Documents\CorsiAulab\Progetto finale\Dataset\emission_dataset(1970-2023)\N2O_Emissions.csv',
    'NOx' : r'C:\Users\Mariapaola\Documents\CorsiAulab\Progetto finale\Dataset\emission_dataset(1970-2023)\NOx_Emissions.csv',
    'SO2' : r'C:\Users\Mariapaola\Documents\CorsiAulab\Progetto finale\Dataset\emission_dataset(1970-2023)\SO2_emissions.csv'
}

def load_datasets(dict_paths):
  '''Carica più dataset da un elenco di percorsi.'''
  datasets = {}
  for name, path in dict_paths.items():
    try:
      df = pd.read_csv(path)
      datasets[name] = df
      print(f"Il dataset {name} è stato caricato con successo. Righe: {df.shape[0]}")
    except FileNotFoundError:
      print(f"Il file {path} non è stato trovato.")
  return datasets

# Dizionario contenente i dataframe caricati
df_emissions = load_datasets(datasets)
```
